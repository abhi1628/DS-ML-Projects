## **What is Data Analytics?**
Data Analytics is the process of examining raw data to find useful information, patterns, and insights. It involves cleaning, organizing, and analyzing data to answer specific questions or solve problems. For example:

- A retail store might analyze sales data to find out which products are selling the most.

- A website might analyze user behavior to improve the user experience.

Data Analytics is often focused on descriptive analytics (what happened) and diagnostic analytics (why it happened). It uses tools like Excel, SQL, Tableau, or Power BI to visualize and interpret data.

## **What is Data Science?**
Data Science is a broader field that includes Data Analytics but goes much further. It involves using advanced techniques like machine learning (ML), artificial intelligence (AI), and statistical modeling to predict future outcomes or make decisions.

Data Science involves predictive analytics (what might happen) and prescriptive analytics (what should be done). It requires programming skills (Python, R) and knowledge of algorithms, statistics, and big data tools.

```python
import pandas as pd
```

```python
data = pd.read_csv('/content/iris.csv')
```

```python
data.info()
```

```python
X = data[['sepal_length','sepal_width','petal_length','petal_width']]
y = data['species']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of model is:", accuracy_score(y_test, y_pred))
```

```python
X_new = [[3,4,1,2],[2,1,4,3]]
y_pred = knn.predict(X_new)
print(y_pred)
```
