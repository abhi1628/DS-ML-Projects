## **Super Simple Spam Classifier**

```python
import pandas as pd
```

```python
data = {'label':['mail','spam','mail','spam'],
        'text':['Hey, How are you?',
                'Win a free IPhone',
                'Can we meet tomorrow?',
                'Claim your prize today!!!']
        }
```

```python
data
```

```python
data = pd.DataFrame(data)
data
```

## Convert words to numbers

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
X = v.fit_transform(data['text'])
print('Words tracked:', v.get_feature_names_out())
print('Word counts:\n', X.toarray())
```

## Train the spam detector

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
y = data['label'].map({'mail':0, 'spam':1})
mnb = MultinomialNB()
mnb.fit(X,y)
```

```python
new_msg = 'Free prize!'
new_count = v.transform([new_msg])
pred = mnb.predict(new_count)
print('Spam (1) or Mail (0)?. It is',pred)
```

## **Data Visualization**

```python
import pandas as pd
data = {'fruits':['Apple','Orange', 'Papaya', 'Mango', 'Pears'],
        'votes':[15,12,8,10,5]}
```

```python
data = pd.DataFrame(data)
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(data['votes'], labels = data['fruits'])
plt.show()
```

```python
plt.pie(data['votes'], labels = data['fruits'], autopct = '%1.1f%%')
plt.show()
```

```python
plt.pie(data['votes'], labels = data['fruits'],
        autopct = '%1.1f%%', explode = [0.1,0,0,0,0])
plt.show()
```

```python
plt.pie(data['votes'], labels = data['fruits'],
        autopct = '%1.1f%%', explode = [0.1,0,0,0,0],
        shadow = True)
plt.show()
```

```python
plt.pie(data['votes'], labels = data['fruits'],
        autopct = '%1.1f%%', explode = [0.1,0,0,0,0],
        shadow = True, colors = ['red','yellow','orange','gold','purple'])
plt.show()
```

```python
plt.pie(data['votes'], labels = data['fruits'],
        autopct = '%1.1f%%', explode = [0.1,0,0,0,0],
        shadow = True, colors = ['red','yellow','orange','gold','purple'])
plt.title('Favorite Fruit Survey')
plt.show()
```
